```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = pd.read_csv('C:\\Users\\USER\\Desktop\\data sheets\\datafile.csv')
```

```python
print(flights)
```

```python
NEW_flights = flights.dropna(how='any')
NEW_flights = flights.dropna(how='any')
df_first_40 = NEW_flights.head(40)
```

```python
df_dropped = df_first_40.drop(columns=['CHLOC1CH','CHLOC1TH','CHRMINH','CHAORD'])
df_dropped.head(40)
```

```python
df_renamed = df_dropped.rename(columns={'CHLOCCT': 'country', 'CHLOC1T': 'city','CHTERM':'Terminal','CHOPERD':'The Airline','CHLOC1D':'airport','CHLOC1':'Airport code'
,'CHRMINE':'flight status' ,'CHCKZN':'Check-in area','CHCINT':'boarding gate','CHPTOL':'Actual Time of Departure' ,'CHSTOL':'Time of Departure'
,'CHFLTN':'Flight num','CHOPER':'Airline code'})
df_renamed.index = range(1, len(df_renamed) + 1)
df_renamed.head(40)
```

```python
flight_times = []
for city in df_renamed['city']:
    if city == 'NEW YORK':
        flight_times.append(12)
    elif city == 'PARIS' or city == 'BARCELONA':
        flight_times.append(4.45)
    elif city == 'MADRID':
        flight_times.append(5.2)
    elif city == 'ROME' or city == 'VIENNA':
        flight_times.append(3.45)
    elif city in ['BUCHAREST', 'BOURGAS', 'VARNA', 'KEFALONIA', 'SOCHI']:
        flight_times.append(2.45)
    elif city in[ 'RAMON'  ,'PAPHOS' , 'LARNACA']:
        flight_times.append(1)
    elif city in ['BUDAPEST', 'DUBAI', 'TBILISI']:
        flight_times.append(3.3)
    elif city in ['ZURICH', 'MILAN', 'PRAGUE']:
        flight_times.append(4.15)
    elif city == 'ATHENS' or city == 'HERAKLION':
        flight_times.append(1.5)
    elif city == 'IASI':
        flight_times.append(3.05)
    else:
        flight_times.append(2.3)
df_renamed['Flight Time (h)'] = flight_times
df_renamed.head(40)
```

```python
df_renamed['Time of Departure'] = pd.to_datetime(df_renamed['Time of Departure'])
df_renamed['Actual Time of Departure'] = pd.to_datetime(df_renamed['Actual Time of Departure'])
df_renamed['Departure Delay'] =df_renamed['Actual Time of Departure'] - df_renamed['Time of Departure']
df_renamed['Early Departure'] = df_renamed['Departure Delay'].apply(lambda x: 'Yes' if x < pd.Timedelta(0) else 'No')
df_renamed.head(40)
```

```python
flightt_counts = df_renamed['Terminal'].value_counts()
print(flightt_counts)
plt.figure(figsize=(8, 8))
flightt_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Number of Flights per Terminal')
plt.ylabel('') 
plt.show()
```

```python
checkin_counts = df_renamed['Check-in area'].value_counts()
labels = [f'{area}: {count}' for area, count in checkin_counts.items()]
plt.figure(figsize=(10, 6))
plt.pie(checkin_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Number of Flights by Check-in Area')
plt.axis('equal')
plt.show()
```

```python
gate_counts = df_renamed['boarding gate'].value_counts()
labels = [f'{gate}: {count}' for gate, count in gate_counts.items()]
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(gate_counts, labels=gate_counts.index, autopct='', startangle=140, wedgeprops=dict(width=0.3))
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Number of Flights by Boarding Gate Area')
labels = [f'{gate}: {count}' for gate, count in gate_counts.items()]
plt.legend(wedges, labels, title="Boarding Gate Area", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
for autotext in autotexts:
    autotext.set_text('')
plt.show()
```

```python
flight_counts = df_renamed['country'].value_counts()

print("\nNumber of flights to each destination:")
print(flight_counts)

plt.figure(figsize=(10, 6))
flight_counts.plot(kind='line', marker='o')
plt.title('Number of Flights to Each Destination')
plt.xlabel('Country')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
flight_countsc = df_renamed['city'].value_counts()
print("\nNumber of flights to each destination:")
flight_countsc.head(40)
```

```python

top_3_destinations = flight_countsc.head(3)

bottom_3_destinations = flight_countsc.tail(3)


combined = pd.concat([top_3_destinations, bottom_3_destinations])


plt.figure(figsize=(10, 6))
combined.plot(kind='bar', color='skyblue')
plt.title('Top 3 and Bottom 3 Flight Destinations')
plt.xlabel('City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

airline_counts2 = df_renamed['Airline code'].value_counts()

plt.figure(figsize=(12, 8))
airline_counts2.plot(kind='bar', color='skyblue')
plt.title('Number of Flights by Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_renamed['Departure Delay'] = (df_renamed['Actual Time of Departure'] - df_renamed['Time of Departure']).dt.total_seconds() / 60
average_delay = df_renamed['Departure Delay'].mean()
average_delay_by_airline = df_renamed.groupby('The Airline')['Departure Delay'].mean()
print(average_delay)
print(average_delay_by_airline)
```

```python
plt.figure(figsize=(12, 8))
bars = average_delay_by_airline.plot(kind='bar', color='skyblue')
plt.axhline(average_delay, color='red', linestyle='--', label=f'Overall Average Delay: {average_delay:.2f} minutes')
plt.title('Average Departure Delay of Each Airline')
plt.xlabel('Airline')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(rotation=45, ha='right')
for i, (airline, avg_delay) in enumerate(average_delay_by_airline.items()):
    plt.text(i, avg_delay + 1, f'{avg_delay:.2f}', ha='center', va='bottom')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
conditions = [
    (df_renamed['Flight Time (h)'] > 4),
    (df_renamed['Flight Time (h)'] > 3) & (df_renamed['Flight Time (h)'] <= 4),
    (df_renamed['Flight Time (h)'] >= 1) & (df_renamed['Flight Time (h)'] <= 2),
    (df_renamed['Flight Time (h)'] < 1)
]

categories = ['Over 4 hours', '3 to 4 hours', '1 to 2 hours', 'Less than 1 hour']
df_renamed['Flight Duration Category'] = pd.cut(df_renamed['Flight Time (h)'], bins=[0, 1, 2, 3, float('inf')], labels=['Less than 1 hour', '1 to 2 hours', '3 to 4 hours', 'Over 4 hours'])
flight_counts = df_renamed['Flight Duration Category'].value_counts().reindex(categories)

print("\nNumber of flights in each category:")
print(flight_counts)
```

```python
plt.figure(figsize=(12, 8))
flight_counts.plot(kind='bar')
plt.title('Number of Flights by Duration Category')
plt.xlabel('Flight Duration')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
